```python
# Лабораторная работа 3
# Метод фиктивного розыгрыша Брауна–Робинсона
# Вариант 3
# N - количество итераций
N = 100
#A = [[1, 9, 6, 0], [-2, 3, 8, 4], [-5, -2, 10, -3], [7, 4, -2, -5]]
A = [[6,1,4],[2,4,2],[4,3,5]]
INF = 99999999
 
# Начальные стратери (произвольное):
P = [1.0, 0.0, 0.0]
Q = [1.0, 0.0, 0.0]
alpha = 0
beta = 0
 
 
def calcNextI(P, Q):
    maxIndex = 0
    maxSum = 0
    for i in range(0, len(A)):
        sum = 0
        for j in range(0, len(A[0])):
            sum += A[i][j] * Q[j]
        if sum > maxSum:
            maxSum = sum
            maxIndex = i
    return (maxIndex, maxSum)
            
def calcNextJ(P, Q):
    minIndex = 0
    minSum = INF
    for j in range(0, len(A[0])):
        sum = 0
        for i in range(0, len(A)):
            sum += A[i][j] * P[i]
        if sum < minSum:
            minSum = sum
            minIndex = j
    return (minIndex, minSum)
 
def solveProblem(A, N):
    
    for step in range(2, N + 1):
        
        # Выбираем новые чистые стратегии 
        nextI, alpha = calcNextI(P, Q)
        nextJ, beta = calcNextJ(P, Q)
        
        # Пересчет частот стратегий
        for i in range(0, len(P)):
            if i != nextI:
                P[i] = (step - 1) * P[i] / step
            else:
                P[i] = ((step - 1) * P[i] + 1) / step
        for j in range(0, len(Q)):
            if j != nextJ:
                Q[j] = (step - 1) * Q[j] / step
            else:
                Q[j] = ((step - 1) * Q[j] + 1) / step
        print ('v(' + str(step) + ') =', (alpha + beta)/2)
        
    print ('P =', P)
    print ('Q = ', Q)
    print ('v =', (alpha + beta)/2)
        
solveProblem(A, N)
```
